# Tutorial: Static Analysis (Band Structure & Wavefunctions)

This notebook demonstrates how to analyze the static properties of 1D topological insulator models using the `hhg` package.

We will cover:
1.  **Band Structure Plotting**: Visualizing the energy dispersion in k-space.
2.  **Wavefunction Plotting**: Visualizing the spatial distribution of eigenstates.


```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt

if os.getcwd() not in sys.path:
    sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from hhg.models import SSHModel
from hhg.analysis import plot_band_structure, plot_wavefunction
```

## 1. Band Structure Analysis

The `plot_band_structure` function diagonalizes the finite chain Hamiltonian and performs a Fourier Transform (FFT) on the eigenvectors to recover the k-space spectral weight.


```python
# Initialize Models
ssh_trivial = SSHModel(N=800, delta=0.15)
ssh_topo = SSHModel(N=800, delta=-0.15)

# Plot SSH Trivial
plt.figure(figsize=(6, 5))
ax = plt.gca()
plot_band_structure(ssh_trivial, label="SSH Trivial Phase", ax=ax)
plt.show()

# Plot SSH Topological
plt.figure(figsize=(6, 5))
ax = plt.gca()
plot_band_structure(ssh_topo, label="SSH Topological Phase", ax=ax)
plt.show()
```

## 2. Wavefunction Plotting

The `plot_wavefunction` function allows you to inspect specific eigenstates. This is particularly useful for identifying **Edge States** in topological phases.


```python
# 2.1 Plot Wavefunction Amplitude
plt.figure(figsize=(8, 4))
plot_wavefunction(ssh_topo, states=[399, 400], kind="wavefunction", title="SSH Edge States (Amplitude)")
plt.show()

# 2.2 Plot Probability Density |psi|^2
plt.figure(figsize=(8, 4))
plot_wavefunction(ssh_topo, states=[399, 400], kind="density", title="SSH Edge States (Density)")
plt.show()

# 2.3 Plotting Bulk States
plt.figure(figsize=(8, 4))
plot_wavefunction(ssh_topo, states=range(398, 401), kind="wavefunction", title="SSH Bulk States")
plt.show()
```
